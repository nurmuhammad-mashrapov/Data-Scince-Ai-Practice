```python
df = pd.read_csv("./Practicum_Datasets/world_cities_air_water_quality.csv")
df
```

```python
df.info()
```

```python
columns = ['City', 'Region', 'Country', 'AirQuality', 'WaterPollution']
df.columns = columns
df
```

```python
df['Region'] = df['Region'].map(lambda x: x.replace('"', "")) # Region ustuniga tegishli ma'lumotlardan " belgisini olib tashlash
df
```

```python
df['Country'] = df['Country'].map(lambda x: x.replace('"', "")) # Country ustuniga tegishli ma'lumotlardan " belgisini olib tashlash
df
```

```python
df.to_csv("Updated_dataset.csv", index=False)
```

```python
df['Country'] = df['Country'].map(lambda x: x.strip()) # Country ustuniga tegishli ma'lumotlardan bo'sh belgi(whitespace)ni belgisini olib tashlash
df
```
